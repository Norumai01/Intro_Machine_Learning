```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets
from torchvision import transforms
import time
```

```python
# Check if a GPU is available
if torch.cuda.is_available():
    # Create a device object for the GPU
    device = torch.device('cuda')
else:
    # Create a device object for the CPU
    device = torch.device('cpu')
device
```

```python
cifar10 = datasets.CIFAR10('data', train=True, download=True)
cifar10_val = datasets.CIFAR10('data', train=False, download=True)
```

```python
class_names = ['airplane','automobile','bird','cat','deer',
               'dog','frog','horse','ship','truck']

fig = plt.figure(figsize=(8,3))
num_classes = 10
for i in range(num_classes):
    ax = fig.add_subplot(2, 5, 1 + i, xticks=[], yticks=[])
    ax.set_title(class_names[i])
    img = next(img for img, label in cifar10 if label == i)
    plt.imshow(img)
plt.show()
```

```python
cifar10 = datasets.CIFAR10('data', train=True, download=False,
                          transform=transforms.Compose([
                              transforms.ToTensor(),
                              transforms.Normalize((0.4915, 0.4823, 0.4468),
                                                   (0.2470, 0.2435, 0.2616))
                          ]))
```

```python
cifar10_val = datasets.CIFAR10('data', train=False, download=False,
                          transform=transforms.Compose([
                              transforms.ToTensor(),
                              transforms.Normalize((0.4915, 0.4823, 0.4468),
                                                   (0.2470, 0.2435, 0.2616))
                          ]))
```

```python
# Problem 1a
start_time = time.time()
train_loader = torch.utils.data.DataLoader(cifar10, batch_size=64, shuffle=True)

model_onelayer = nn.Sequential(
            nn.Linear(3072, 128),
            nn.ReLU(),
            nn.Linear(128, 10))

model_onelayer = model_onelayer.to(device)

learning_rate = 1e-3

optimizer = optim.SGD(model_onelayer.parameters(), lr=learning_rate)

loss_fn = nn.CrossEntropyLoss()

n_epochs = 300

one_epochs = []
one_loss = []

for epoch in range(n_epochs):
    for imgs, labels in train_loader:
        imgs = imgs.to(device)
        labels = labels.to(device)
        outputs = model_onelayer(imgs.view(imgs.shape[0], -1))
        loss = loss_fn(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    one_epochs.append(epoch)
    one_loss.append(float(loss))
    if epoch % 30 == 0 or epoch == 299:
        print("Epoch: %d, Loss: %f" % (epoch, float(loss)))
end_time = time.time()
elapsed = end_time - start_time
print("Training Time was:", elapsed, "seconds.")
```

```python
# Printing training accuracy
train_loader = torch.utils.data.DataLoader(cifar10, batch_size=64, shuffle=False)

correct = 0
total = 0

with torch.no_grad():
    for imgs, labels in train_loader:
        imgs = imgs.to(device)
        labels = labels.to(device)
        outputs = model_onelayer(imgs.view(imgs.shape[0], -1))
        _, predicted = torch.max(outputs, dim=1)
        total += labels.shape[0]
        correct += int((predicted == labels).sum())
        
print("Training Accuracy: %f" % (correct / total))

# Printing Validation accuracy
val_loader = torch.utils.data.DataLoader(cifar10_val, batch_size=64, shuffle=False)

correct = 0
total = 0

with torch.no_grad():
    for imgs, labels in val_loader:
        imgs = imgs.to(device)
        labels = labels.to(device)
        outputs = model_onelayer(imgs.view(imgs.shape[0], -1))
        _, predicted = torch.max(outputs, dim=1)
        total += labels.shape[0]
        correct += int((predicted == labels).sum())
        
print("Validation Accuracy: %f" % (correct / total))
```

```python
plt.figure(1)
plt.plot(one_epochs, one_loss)
plt.xlabel("Number of epochs")
plt.ylabel("Loss")
plt.title("Number of epochs vs. Loss")
plt.show()
```

```python
cifar10 = datasets.CIFAR10('data', train=True, download=False,
                          transform=transforms.Compose([
                              transforms.ToTensor(),
                              transforms.Normalize((0.4915, 0.4823, 0.4468),
                                                   (0.2470, 0.2435, 0.2616))
                          ]))
cifar10_val = datasets.CIFAR10('data', train=False, download=False,
                          transform=transforms.Compose([
                              transforms.ToTensor(),
                              transforms.Normalize((0.4915, 0.4823, 0.4468),
                                                   (0.2470, 0.2435, 0.2616))
                          ]))
```

```python
# Problem 1b
start_time = time.time()
train_loader = torch.utils.data.DataLoader(cifar10, batch_size=64, shuffle=True)

model_threelayer = nn.Sequential(
            nn.Linear(3072, 1024),
            nn.ReLU(),
            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.Linear(512, 128),
            nn.ReLU(),
            nn.Linear(128, 10))

model_threelayer = model_threelayer.to(torch.device('cuda'))

learning_rate = 1e-3

optimizer = optim.SGD(model_threelayer.parameters(), lr=learning_rate)

loss_fn = nn.CrossEntropyLoss()

n_epochs = 300

three_epochs = []
three_loss = []

for epoch in range(n_epochs):
    for imgs, labels in train_loader:
        imgs = imgs.to(torch.device('cuda'))
        labels = labels.to(torch.device('cuda'))
        outputs = model_threelayer(imgs.view(imgs.shape[0], -1))
        loss = loss_fn(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    three_epochs.append(epoch)
    three_loss.append(float(loss))
    if epoch % 30 == 0 or epoch == 299:
        print("Epoch: %d, Loss: %f" % (epoch, float(loss)))
end_time = time.time()
elapsed = end_time - start_time
print("Training Time was:", elapsed, "seconds.")
```

```python
# Printing training accuracy
train_loader = torch.utils.data.DataLoader(cifar10, batch_size=64, shuffle=False)

correct = 0
total = 0

with torch.no_grad():
    for imgs, labels in train_loader:
        imgs = imgs.to(torch.device('cuda'))
        labels = labels.to(torch.device('cuda'))
        outputs = model_threelayer(imgs.view(imgs.shape[0], -1))
        _, predicted = torch.max(outputs, dim=1)
        total += labels.shape[0]
        correct += int((predicted == labels).sum())
        
print("Training Accuracy: %f" % (correct / total))

# Printing Validation accuracy
val_loader = torch.utils.data.DataLoader(cifar10_val, batch_size=64, shuffle=False)

correct = 0
total = 0

with torch.no_grad():
    for imgs, labels in val_loader:
        imgs = imgs.to(torch.device('cuda'))
        labels = labels.to(torch.device('cuda'))
        outputs = model_threelayer(imgs.view(imgs.shape[0], -1))
        _, predicted = torch.max(outputs, dim=1)
        total += labels.shape[0]
        correct += int((predicted == labels).sum())
        
print("Validation Accuracy: %f" % (correct / total))
```

```python
plt.figure(2)
plt.plot(three_epochs, three_loss)
plt.xlabel("Number of epochs")
plt.ylabel("Loss")
plt.title("Number of epochs vs. Loss")
plt.show()
```
